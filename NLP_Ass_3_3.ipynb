```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4') # Open Multilingual Wordnet, often needed with wordnet

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN # Default to noun if more specific tag not found

nltk_lemmas = []
for word, tag in nltk_pos_tags:
    wn_tag = get_wordnet_pos(tag)
    nltk_lemmas.append(lemmatizer.lemmatize(word, wn_tag))

spacy_lemmas = [token.lemma_ for token in doc]

print("--- NLTK Lemmas (first 10) ---")
print(nltk_lemmas[:10])

print("\n--- spaCy Lemmas (first 10) ---")
print(spacy_lemmas[:10])
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

nltk_stems = [stemmer.stem(word) for word in nltk_word_tokens]
spacy_lemmas = [token.lemma_ for token in doc] # Re-using spaCy lemmas for comparison

print("--- NLTK Stems (first 10) ---")
print(nltk_stems[:10])

print("\n--- spaCy Lemmas (first 10) ---")
print(spacy_lemmas[:10])
```

```python
tag_differences = []
# Assuming nltk_pos_tags and spacy_pos_tags are aligned by word order
# However, NLTK tags might be slightly different in length due to tokenization differences (e.g., punctuation attached)
# Let's try to align them by word text for robust comparison, taking the shorter length for safety.
min_len = min(len(nltk_pos_tags), len(spacy_pos_tags))

for i in range(min_len):
    nltk_word, nltk_tag = nltk_pos_tags[i]
    spacy_word, spacy_tag = spacy_pos_tags[i]

    # Compare only if the words are approximately the same (case-insensitive) to account for slight tokenization variations
    if nltk_word.lower() == spacy_word.lower():
        if nltk_tag != spacy_tag:
            tag_differences.append((nltk_word, nltk_tag, spacy_tag))

print("--- POS Tag Differences (NLTK vs. spaCy) ---")
if tag_differences:
    for diff in tag_differences[:10]: # Print first 10 differences
        print(f"Word: '{diff[0]}', NLTK Tag: '{diff[1]}', spaCy Tag: '{diff[2]}'")
    if len(tag_differences) > 10:
        print(f"... and {len(tag_differences) - 10} more differences.")
else:
    print("No significant tag differences observed for aligned words.")
```

```python
extracted_nouns = []
extracted_verbs = []

for word, pos_tag in spacy_pos_tags:
    if pos_tag == 'NOUN':
        extracted_nouns.append(word)
    elif pos_tag == 'VERB':
        extracted_verbs.append(word)

print("--- Extracted Nouns (first 10) ---")
print(extracted_nouns[:10])

print("\n--- Extracted Verbs (first 10) ---")
print(extracted_verbs[:10])
```

```python
from collections import Counter

noun_frequencies = Counter(extracted_nouns)
verb_frequencies = Counter(extracted_verbs)

print("--- Top 10 Most Common Nouns ---")
for noun, count in noun_frequencies.most_common(10):
    print(f"{noun}: {count}")

print("\n--- Top 10 Most Common Verbs ---")
for verb, count in verb_frequencies.most_common(10):
    print(f"{verb}: {count}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for plotting
top_10_nouns = noun_frequencies.most_common(10)
top_10_verbs = verb_frequencies.most_common(10)

noun_labels, noun_counts = zip(*top_10_nouns)
verb_labels, verb_counts = zip(*top_10_verbs)

# Create a figure with two subplots
plt.figure(figsize=(16, 6)) # Adjust figure size for better readability

# Subplot for Nouns
plt.subplot(1, 2, 1) # 1 row, 2 columns, first subplot
sns.barplot(x=list(noun_labels), y=list(noun_counts), palette='viridis')
plt.title('Top 10 Most Frequent Nouns')
plt.xlabel('Noun')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability

# Subplot for Verbs
plt.subplot(1, 2, 2) # 1 row, 2 columns, second subplot
sns.barplot(x=list(verb_labels), y=list(verb_counts), palette='magma')
plt.title('Top 10 Most Frequent Verbs')
plt.xlabel('Verb')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability

# Adjust layout and display plots
plt.tight_layout()
plt.show()
```
